```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("C:/Users/blaqj/Downloads/real_estate_price_size (1).csv")
df
```

```python
x = df['size']
y = df['price']
```

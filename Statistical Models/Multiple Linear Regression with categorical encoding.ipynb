```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv('C:/Users/blaqj/Downloads/Data Science Course/1.03.+Dummies.csv')
df
```

```python
data = df.copy()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1, 'No':0})
```

```python
data
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
```

```python
results.summary()
```

```python
x
```

```python
new_data = pd.DataFrame({'count': 1, 'SAT': [1700, 1670], 'Attendance': [0, 1]})
new_data = new_data[['count', 'SAT', 'Attendance']]
new_data
```

```python
predictions = results.predict(new_data)
predictions
```

```python
predictionsdf = pd.DataFrame({'Predictions': predictions})
joined = new_data.join(predictionsdf)
joined.rename(index = {0: 'Bob', 1: 'Alice'})
```
